<h1 style="text-align:center; font-size:60px;">Arrecifes de coral</h1>

<p style="text-align: justify; font-size:34px;">
Los arrecifes de coral son uno de los ecosistemas más diversos del planeta. 
Se distribuyen en menos del 1% del océano; sin embargo, albergan cerca de la cuarta parte de toda la biodiversidad marina.
</p>

<p style="font-size: 30px;"><strong>Importancia:</strong></p>

<ul style="font-size:30px;">
  <li>Constituyen áreas vitales de refugio y reproducción para numerosas especies.</li>
  <li>Protegen las costas frente a la erosión causada por el oleaje.</li>
  <li>Son fuente de recursos pesqueros como langosta, pargos y meros.</li>
  <li>Contribuyen a la formación de arena que nutre las playas.</li>
</ul>

<div style="background:orange; color:white; padding:20px; border-radius:10px;">
<h1 style='text-align:center; font-size:50px;' >En México . . .</h1>
<p style="font-size: 20px;">En México, estos ecosistemas se distribuyen en tres regiones principales: la costa norte del Pacífico, la
costa de Veracruz y Campeche, y la costa oriental de la península de Yucatán.</p>

</div>
<div style="font-size:12px;">
Superficie de arrecifes expresada en hectáreas (ha) con base en datos de CONABIO.
</div>



```python
import plotly.io as pio
pio.renderers.default = "iframe"
import plotly.express as px
import pandas as pd

# Datos con especies
data = {
    "Lugar": ["Sistema Arrecifal Veracruzano", "Sistema Arrecifal Lobos -Tuxpan", "Laguna de Términos", "Arrecife Alacranes", "Arrecife de Puerto Morelos",
             "Arrecifes de Cozumel", "Banco Chinchorro", "Arrecifes de Xcalak", "Arrecifes de Sian Kaan", "Cabo Pulmo", "Bahía de La Paz"],
    "Latitud": [19.2, 20.9, 18., 22.2, 20.4, 20.2, 18.3, 18.1, 19.3, 23.2, 24.0],
    "Longitud": [-96.1, -97.4, -91., -89.4, -86.5, -86.5, -87.1, -87.5, -87.3, -109.2, -110.2],
    "Superficie total": [65516.47, 30571.15, 706147.67, 333768.51, 9066.63, 11987.88, 	
144360.00, 17949.45, 34927.16, 	
7111.01, 2512.73  ]  # valores aproximados
}

df = pd.DataFrame(data)

# Mapa interactivo
fig = px.scatter_geo(
    df,
    lat="Latitud",
    lon="Longitud",
    hover_name="Lugar",
    hover_data={"Superficie total": True},  # 👈 esto agrega la info
    title="Ubicación de Arrecifes en México"
)
fig.update_traces(marker=dict(color="DeepPink"))
fig.update_geos(
    center=dict(lat=20, lon=-95),
    projection_scale=5
)

fig.show()

```

##

<div style="background:limegreen; color:white; padding:20px; border-radius:10px;">
<h1 style='text-align:center; font-size:50px; '>Deterioro</h1>
<p style="font-size: 20px;">Los arrecifes de coral están experimentando una disminución significativa en su cobertura a nivel global. 
Este deterioro ocurre como resultado de múltiples factores como el aumento de la temperatura del océano, la contaminación y la actividad humana.</p>
<p style="font-size: 20px;">En la gráfica se muestra una estimación basada en los reportes del 
Global Coral Reef Monitoring Network</p>
</div>



```python
import matplotlib.pyplot as plt

years = [1985, 2000, 2010, 2020]
coral_cover = [50, 40, 30, 20]  # % aproximado basado en tendencias globales

plt.plot(years, coral_cover, marker='o')
plt.xlabel("Año")
plt.ylabel("Cobertura coralina (%)")
plt.title("Deterioro de arrecifes de coral")
plt.show()
```

<div style="font-size:12px;">
Estimaciones basadas en reportes del GCRMN sobre pérdida de cobertura coralina.
</div>


##

<div style="background:dodgerblue; color:white; padding:20px; border-radius:10px;">
<h1 style='text-align:center; font-size:60px; '>¿Por qué se están perdiendo?</h1>
<p style="font-size: 30px;">
Principalmente, por la gran diversidad de fauna y flora de los arrecifes coralinos se generan paisajes submarinos de gran valor estético, esto los convierte en lugares con alta demanda turística internacional, resultando un recurso natural explotable económicamente.
Las especies presentes en los arrecifesson fuente de sustancias que se emplean como fármacos a partir de sus propiedades y son utilizados como reactivos de interés bioquímico y experimental.
Además, al se fuente de una gran variedad de recursos pesqueros, tiende a la sobrepesca y al empleamiento de técnicas inadecuadas que destruyen los corales.</p>
</div>

<h1 style='text-align:center; font-size:60px; '>¿Nadaremos?</h1>
<p style="font-size: 30px;">
Los arrecifes de coral son fundamentales para la biodiversidad marina y el equilibrio de los ecosistemas oceánicos. 
Sin embargo, su deterioro evidencia el impacto directo de las actividades humanas sobre el medio ambiente.
A través del análisis de su ubicación y la disminución de su cobertura, es posible comprender que la crisis de los arrecifes no es un problema aislado, sino una consecuencia de un modelo de consumo y uso de recursos que no es sostenible.
En el marco del Objetivo de Desarrollo Sostenible 14, proteger estos ecosistemas implica no solo conservar la vida submarina, sino también replantear la relación entre las acciones humanas y el océano.</p>
